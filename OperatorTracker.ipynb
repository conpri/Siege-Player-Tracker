```python
from ultralytics import YOLO

model = YOLO("yolo26n.pt")  # start small
```

```python
import cv2
import matplotlib.pyplot as plt
import os

img_dir = "Siege Dataset/Siege Dataset/train/images"
lbl_dir = "Siege Dataset/Siege Dataset/train/labels"

img_name = os.listdir(img_dir)[0]
img = cv2.imread(os.path.join(img_dir, img_name))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

h, w, _ = img.shape

with open(os.path.join(lbl_dir, img_name.replace(".png", ".txt"))) as f:
    for line in f:
        cls, xc, yc, bw, bh = map(float, line.split())
        x1 = int((xc - bw / 2) * w)
        y1 = int((yc - bh / 2) * h)
        x2 = int((xc + bw / 2) * w)
        y2 = int((yc + bh / 2) * h)
        cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 0), 2)

plt.imshow(img)
plt.axis("off")
```

```python
model.train(
    data="Siege Dataset/Siege Dataset/data.yaml",
    device = 0
)
```
